```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Generate some synthetic data for demonstration
np.random.seed(0)
distance = np.random.uniform(1, 100, 100)  # Generating random distances
noise = np.random.normal(0, 5, 100)  # Adding some random noise
signal_strength = 100 - 0.5 * distance + noise  # Generating signal strength based on distance with noise

# Reshape the data
distance = distance.reshape(-1, 1)  # Reshaping to a column vector

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(distance, signal_strength, test_size=0.2, random_state=42)

# Create and train the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions
y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)

# Evaluate the model
train_rmse = np.sqrt(mean_squared_error(y_train, y_pred_train))
test_rmse = np.sqrt(mean_squared_error(y_test, y_pred_test))
print(f"Train RMSE: {train_rmse:.2f}")
print(f"Test RMSE: {test_rmse:.2f}")

# Plot the results
plt.figure(figsize=(10, 6))
plt.scatter(X_train, y_train, color='blue', label='Training data')
plt.scatter(X_test, y_test, color='red', label='Testing data')
plt.plot(X_train, y_pred_train, color='green', label='Predictions (Training)')
plt.plot(X_test, y_pred_test, color='orange', label='Predictions (Testing)')
plt.xlabel('Distance')
plt.ylabel('Signal Strength')
plt.title('Linear Regression for Signal Strength Prediction')
plt.legend()
plt.grid(True)
plt.show()
```
